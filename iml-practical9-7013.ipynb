```python
Write a Pandas program to create a plot of Open, High, Low, Close, 
Adjusted Closing prices and Volume of given company between two 
specific dates.
```

```python
import plotly.graph_objects as go
import pandas as pd
from datetime import datetime
```

```python
df=pd.read_csv('/kaggle/input/national-stock-exchange-time-series/infy_stock.csv')
```

```python
fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])

fig.show()
```
