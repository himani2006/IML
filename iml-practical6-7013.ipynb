* to create the mean and standard deviation of the data of a given Series 

```python
import pandas as pd
```

```python
data=pd.Series([1,2,3,4,5,6,7,8,9])
print(data)
print("MEAN DEVISTION=",np.mean(data))
print("STANDERS DEVISTION=",np.std(data))
```
