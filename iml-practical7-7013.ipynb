Write a Pandas program to implement following operation 
* to create a dataframe from a dictionary and display it 

```python
import pandas as pd
import numpy as np
```

```python
dic={
   'Name': ['Himani','Riya','Jiya','Viddhi','Diya'], 
   'Enrollment No.': [13, 3, 6,67,2],
   'Age' :[18,17,17,18,18]
}
```

```python
df=pd.DataFrame(dic)
print("DataFrame =\n",df)
```

* to sort the DataFrame first by 'name' in ascending order 

```python
print("DataFrame befor Shorting Name: \n",df)
df_new=df.sort_values(by=['Name'], ascending=True) 
print("DataFrame after Shorting Name: \n",df_new)
```

* to delete the one specific column from the DataFrame 

```python
del df_new['Name']
df_new
```

* to write a DataFrame to CSV file using tab separator

```python
df_new.to_csv('/kaggle/working/dataset.csv', index=False, sep="\t")
```
